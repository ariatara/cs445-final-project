```python
import os
```

```python
folder = "/mnt/c/users/habib/projects/cs445/final_project/archive"
files = os.listdir(folder)
```

```python
for f in files:
    if ("data" in f) or ("Readme" in f) or ('.png' in f): continue
    else: 
        old = os.path.join(folder, f)
        new_path = os.path.join(folder, f.replace(".","_")+".gif")
        os.rename(old, new_path)
```

```python
#after moving all gifs to images folder
from PIL import Image
folder = "images"
files = os.listdir(folder)
for f in files:
    if ('.gif' not in f): continue
    im_file = os.path.join(folder, f)
    im = Image.open(im_file)
    im.save(im_file.replace(".gif",".png")) 
```
